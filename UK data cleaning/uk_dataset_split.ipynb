# uk dataset split
takes the labeled parlvote csv and splits into train/val/test
also filters out very short speeches

```python
import pandas as pd
from sklearn.model_selection import train_test_split
INPUT_FILE= '/kaggle/input/datasets/zainabrizvi5/thesis-data/parlvote_labeled.csv'
OUTPUT_DIR= '/kaggle/working/'
```

```python
df= pd.read_csv(INPUT_FILE)
print(f'loaded {len(df)} rows')
print(df['label'].value_counts())
```

```python
# cutting the very short speeches, less than 20 words  not useful
df= df[df['word_count'] >= 20].reset_index(drop=True)
print(f'after filtering short speeches: {len(df)} rows')
print(df['label'].value_counts())
```

```python
# keep only text and label columns
df= df[['text','label']]
# 70/15/15 split, stratified so class proportions are preserved
train_df, temp_df = train_test_split(df, test_size= 0.30, stratify=df['label'], random_state=42)
val_df, test_df= train_test_split(temp_df, test_size =0.50, stratify=temp_df['label'], random_state=42)
print(f'train: {len(train_df)}, val: {len(val_df)}, test: {len(test_df)}')
```

```python
# check class distribution in each split
for name, split in [('train', train_df), ('val', val_df), ('test', test_df)]:
    print(f'{name}:')
    print(split['label'].value_counts())
    print()
```

```python
train_df.to_csv(OUTPUT_DIR + 'uk_train.csv', index=False)
val_df.to_csv(OUTPUT_DIR + 'uk_val.csv', index=False)
test_df.to_csv(OUTPUT_DIR + 'uk_test.csv', index=False)
print('saved uk_train.csv, uk_val.csv, uk_test.csv')
```
