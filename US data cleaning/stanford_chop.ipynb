# stanford chop
filters to 1997-2016 (congress 105+) to align with parlvote
then downsamples to 67k with equal left/right split

```python
import pandas as pd
df= pd.read_csv('/kaggle/input/datasets/zainabrizvi5/thesis-data/stanford_labeled.csv')
print(f'loaded {len(df):,} rows')
print(df['label'].value_counts())
```

```python
# congress 105 starts in 1997, which is when parlvote begins
df_filtered= df[df['congress'] >= 105].reset_index(drop=True)
print(f'after year filter: {len(df_filtered):,}')
print(df_filtered['label'].value_counts())
```

```python
# downsample to 33500 per class = 67k total, balanced 50/50
df_sampled= df_filtered.groupby('label', group_keys=False).apply(
    lambda x: x.sample(n=33500, random_state=42)
).reset_index(drop=True)

print(f'after downsampling: {len(df_sampled):,}')
print(df_sampled['label'].value_counts())

df_sampled.to_csv('stanford_labeled_final.csv', index=False)
print('saved to stanford_labeled_final.csv')
```
