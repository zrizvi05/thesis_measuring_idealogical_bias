# us dataset split
70/15/15 train/val/test split on the final balanced stanford dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
df= pd.read_csv('/kaggle/input/datasets/zainabrizvi5/thesis-data/stanford_labeled_final.csv')
print(f'loaded {len(df):,} rows')
print(df['label'].value_counts())
```

```python
# stratified split so left/right proportions stay consistent across splits
train, temp= train_test_split(df, test_size=0.30, stratify=df['label'], random_state=42)
val, test = train_test_split(temp, test_size=0.50, stratify=temp['label'], random_state=42)
print(f'train: {len(train):,}, val: {len(val):,}, test: {len(test):,}')
```

```python
# cofirm that distributions look equal
for name, split in [('train', train), ('val', val), ('test', test)]:
    print(f'{name}:')
    print(split['label'].value_counts())
    print()
```

```python
train.to_csv('us_train.csv', index=False)
val.to_csv('us_val.csv', index=False)
test.to_csv('us_test.csv', index=False)
print('saved us_train.csv, us_val.csv, us_test.csv')
```
